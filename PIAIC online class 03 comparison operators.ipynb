Comparison Operators

```python

# Comparison Operators:

# == : Equal to
# != : Not Equal to
# > : Greater than
# < : Less than
# >= : Greater than or equal to
# <= : Less than or equal to

a=90
b=30

print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
```

```python
my_age=int(input("What is your age ?"))
if ( my_age < 18 ) :
    print("You're minor")
elif (my_age==18 ) :
    print ("You're a young person")
elif (my_age>80 ) :
    print ("You're old")
else:
  print("You're an adult")
```

```python
my_age=int(input("What is your age ?"))
weight = int(input("What is your weight ?"))
if ( my_age < 18 ) :
    print("You're minor")
elif (my_age==18 and weight <=80 ) :
    print ("You're a young person")
elif (my_age>80 and weight <=80 ) :
    print ("You're old")
else:
  print("You're Overweight")
```
