If Else

```python
is_hungry = True
burger_lover = False
pizza_lover = False
if(is_hungry and (burger_lover or pizza_lover)):
  print("Lets order the Pizza")
elif(is_hungry):
  print("Let's go to sleep")

else:
  print("Not Hungry")
```

Checking Membership

```python
shopping_items : List[str] = ["Cake", "Eggs", "Juice", "Chips", "Bread", "Oil"]
print("Oil" in shopping_items)
```

Adding or Removing items from list.

```python
shopping_items : List[str] = ["Cake", "Eggs", "Juice", "Chips", "Bread", "Oil"]
print(shopping_items)
# Append item flour
shopping_items.append("Flour")
print(shopping_items)

# Append item flour
shopping_items.append("Tea")
print(shopping_items)


# Remove Append item flour
shopping_items.remove("Oil")
print(shopping_items)




```

Inserting items in List

```python
# Inserting item in list
shopping_items : List[str] = ["Cake", "Eggs", "Juice", "Chips", "Bread", "Oil"]
shopping_items.insert(2,"Tea")
print("After inserting & before popping", shopping_items)

shopping_items.pop(3)
print("After Popping", shopping_items)
```
