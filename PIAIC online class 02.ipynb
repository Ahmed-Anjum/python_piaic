Operator

```python
# Addition
a = 11
b = 5
result = a+b
print ("Addition", result)


# Subtraction
a = 11
b = 5
result = a-b
print ("Subtraction", result)


# Multiplication
a = 11
b = 5
result = a*b
print ("Multiplication", result)


# Division
a = 11
b = 5
result = a/b
print ("Division", result)


# Modulus returns remainder
result = a%b
print ("Modulus", result)


# Power returns remainder (11*11*11*11*11)

result = a**b
print ("Power", result)
```

```python
# Calculate Rabri qty using operators
faculty = 12
admin = 4
sudents = 100
total_member=faculty+admin+sudents
abscent=15
total_member_present=total_member-abscent
print(f"Total Member in PIAIC {total_member}")
print(f"Total Member present in PIAIC {total_member_present}")
rabri_qty_per_person=250
rabri_qty=rabri_qty_per_person*total_member_present
print(f"Total Rabri Qty {rabri_qty} g")
rabrp_qty_in_kg=rabri_qty/1000
print(f"Total Rabri Qtyin kg  {rabrp_qty_in_kg} kg")
```
